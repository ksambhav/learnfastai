```python
import matplotlib.pyplot as plt
import torch

# Define the quadratic function
def quad_fun(x, a, b, c):
    return a * x ** 2 + b * x + c

# Parameters for the quadratic function
a, b, c = 2, 10, 1

# Create x with requires_grad=True to track gradients
x = torch.linspace(-30, 20, 200)[:, None].requires_grad_(True)

# Compute y using the quadratic function directly
y = quad_fun(x, a, b, c)

# Compute the gradient (dy/dx)
dy_dx = torch.autograd.grad(outputs=y, inputs=x, grad_outputs=torch.ones_like(y), create_graph=True)[0]

# Now dy_dx contains the derivative of y with respect to x

# Plotting the function and its derivative for visualization
plt.figure(figsize=(10, 6))
plt.plot(x.detach().numpy(), y.detach().numpy(), label=f'y = ${a}x^2 + {b}x + {c}$')
plt.plot(x.detach().numpy(), dy_dx.detach().numpy(), label='dy/dx', linestyle='--')
plt.legend()
plt.xlabel('x')
plt.ylabel('y / dy/dx')
plt.title('Function and Its Derivative')
plt.show()
```
